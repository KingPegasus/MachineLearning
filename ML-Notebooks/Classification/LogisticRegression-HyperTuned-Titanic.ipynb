### Task 1: Load Data from 'titanic_clean.csv'

```python
#write code here
data = pd.read_csv('datasets/titanic_clean.csv')
df = data.copy()
df.head()
```

```python
print(df.shape)
```

### One Hot encoding for categorical varaibles

```python
df_OneHot=pd.get_dummies(df,columns=['Pclass','Sex','Embarked','Title','GrpSize','FareCat','AgeCat'])
df_OneHot.head()
```

```python
df=df_OneHot.copy()
```

### Task 2: Create Independent and Dependent Variables

```python
#write code here
X = df.drop(columns= ['Survived','PassengerId'], axis =1)
Y = df['Survived']
```

### Task 3: Train Test n Split the data

```python
# Import the library
from sklearn.model_selection import train_test_split
```

```python
#Write the code here
xtrain, xtest, ytrain, ytest = train_test_split(X,Y,test_size=0.3, random_state=25, shuffle = True)
print(xtrain.shape, ytrain.shape)
print(xtest.shape, ytest.shape)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

# Logistic Regression

### Creating Model & Training

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RandomizedSearchCV
```

```python
#solver{‘newton-cg’, ‘lbfgs’, ‘liblinear’, ‘sag’, ‘saga’}
solver = ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
#penalty{‘l1’, ‘l2’, ‘elasticnet’, ‘none’}
penalty = ['l1','l2','elasticnet']
dual =[True, False]
```

```python
random_grid3 = {"solver": solver,
                "penalty": penalty,
               "dual": dual}
```

```python
lr_model = LogisticRegression(random_state=25,max_iter=500)
lr_random = RandomizedSearchCV(
    estimator=lr_model, param_distributions=random_grid3, n_iter=30, cv=5, verbose=0, random_state=25,n_jobs=4)
# Fit the random search model
lr_random.fit(xtrain, ytrain)
lr_random
```

```python
lr_random
```

```python
clf_lr_rand = lr_random.best_estimator_
clf_lr_rand.fit(xtrain, ytrain)
```

```python
pred_clf_lr_rand = clf_lr_rand.predict(xtest)
```

```python
#pred_prb[:,1] #probabilities of 1 (survival)
```

First value in Numpy array is probability of **0** and second is probability of **1**

Only predicting and extracting probability values of **1**

```python
lr_pred_prb_rand = clf_lr_rand.predict_proba(xtest)[:, 1] #probabilities of 1 (survival)
```

```python
from sklearn.metrics import recall_score, accuracy_score, precision_score, f1_score, confusion_matrix
```

```python
lr_random_recall= recall_score(ytest,pred_clf_lr_rand)
lr_random_acc= accuracy_score(ytest,pred_clf_lr_rand)
lr_random_pre= precision_score(ytest,pred_clf_lr_rand)
lr_random_f1_score= f1_score(ytest,pred_clf_lr_rand)

print("Logistic Regression Hyperparameters Tuning Using Randomized Search")
print("Precision: {0:.3f}%".format(lr_random_pre*100))
print("Accuracy:  {0:.3f}%".format(lr_random_acc*100))
print("Recall:    {0:.3f}%".format(lr_random_recall*100))
print("F1-Score:  {0:.3f}%".format(lr_random_f1_score*100))
```

```python
tn, fp, fn, tp = confusion_matrix(ytest, pred_clf_lr_rand).ravel()
conf_matrix=pd.DataFrame({"pred_Survived":[tp,fp],"pred_Not Survived":[fn,tn]},index=["Survived","Not Survived"])
conf_matrix
```
